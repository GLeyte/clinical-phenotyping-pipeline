# Import data

```python
# ===== CARREGAMENTO DOS DADOS =====
data_folder = current_notebook / COMPLETE_DATAPATH

data_covid = pd.read_csv(data_folder / COVID_DATA_FILE)
future_data = pd.read_csv(data_folder / FUTURE_DATA_FILE)
```

```python
comorbidities = [
    "myocardial_infarct", "congestive_heart_failure", "peripheral_vascular_disease", 
    "cerebrovascular_disease", "dementia", "chronic_pulmonary_disease", 
    "rheumatic_disease", "peptic_ulcer_disease", "mild_liver_disease", 
    "diabetes_without_cc", "diabetes_with_cc", "paraplegia", "renal_disease", 
    "malignant_cancer", "severe_liver_disease", "metastatic_solid_tumor", "aids",
]

cats = comorbidities + ["age","died"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute the correlation matrix
correlation_matrix = data_covid[cats].corr()

# Create a figure and set its size
plt.figure(figsize=(15, 15))

# Create a heatmap
sns.heatmap(correlation_matrix, 
            annot=True,           # Show numeric values
            cmap='coolwarm',      # Color palette (red to blue)
            center=0,              # Center color at 0
            vmin=-1, 
            vmax=1,                # Set color scale from -1 to 1
            square=True,           # Make the plot square
            linewidths=0.5,        # Add lines between cells
            cbar_kws={"shrink": .8}) # Slightly reduce colorbar size

# Set the title
plt.title('Correlation COVID patients', fontsize=15)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
