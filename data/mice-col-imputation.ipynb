# Parameters

```python
MAX_ITER = 5
M_IMPUTATION = 10

## INPUT

MODULES_FOLDER = "../"

COMPLETE_DATA_FOLDER = 'mimic_data/complete_data/'

COVID_DATA_FILE = 'patient_covid.csv'
CONTROL_PRE_DATA_FILE = 'patient_control_pre.csv'
CONTROL_POST_DATA_FILE = 'patient_control_post.csv'

## OUTPUT

OUTPUT_DATA_FOLDER = 'mice-col/'

OUTPUT_COVID_TRAIN_DATA_FILE = 'patient_covid_train.csv'
OUTPUT_COVID_TEST_DATA_FILE = 'patient_covid_test.csv'

OUTPUT_CONTROL_PRE_TRAIN_DATA_FILE = 'patient_control_pre_train.csv'
OUTPUT_CONTROL_PRE_TEST_DATA_FILE = 'patient_control_pre_test.csv'

OUTPUT_CONTROL_POST_TRAIN_DATA_FILE = 'patient_control_post_train.csv'
OUTPUT_CONTROL_POST_TEST_DATA_FILE = 'patient_control_post_test.csv'
```

# COVID

## Import all Data

```python
import pandas as pd
import os
import sys

sys.path.append(os.path.abspath(MODULES_FOLDER))

import Modules.ImputationModule as imp

# Set the path to the datasets
current_dir = os.getcwd()
folderpath = os.path.join(current_dir, COMPLETE_DATA_FOLDER)

df = pd.read_csv(folderpath + COVID_DATA_FILE)
```

```python
print(f"Size of df: {df.shape}")
```

## Fill missing values with the MICE

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Generate a sample dataset with missing values
np.random.seed(42)
```

```python
# Split COVIDSET into training and testing sets
train, test = train_test_split(df, test_size=0.3, random_state=42)
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)
```

```python
not_consider = [
    "hadm_id",
    "subject_id",
]
```

```python
mice_1 = imp.MICE_Imputer(m_imputations=M_IMPUTATION, max_iter=MAX_ITER, verbose=True)
_, _, df_train_imputed, df_test_imputed  = mice_1.train_test_impute_with_aggregation(
        df_train=train,
        df_test=test,
        not_consider=not_consider
    )
```

```python
# Check results
print("\n" + "="*60)
print("IMPUTATION SUMMARY")
print("="*60)

# Columns that were imputed
imputed_cols = [col for col in train.columns if col]

# Training set summary
print(f"\nTRAINING SET:")
print(f"Shape: {train.shape} -> {df_train_imputed.shape}")
missing_before_train = train[imputed_cols].isnull().sum().sum()
missing_after_train = df_train_imputed[imputed_cols].isnull().sum().sum()
print(f"Missing values: {missing_before_train} -> {missing_after_train}")

# Test set summary  
print(f"\nTEST SET:")
print(f"Shape: {test.shape} -> {df_test_imputed.shape}")
missing_before_test = test[imputed_cols].isnull().sum().sum()
missing_after_test = df_test_imputed[imputed_cols].isnull().sum().sum()
print(f"Missing values: {missing_before_test} -> {missing_after_test}")
```

### Separate data

```python
covid_train_imputed = df_train_imputed[df_train_imputed['hadm_id'].isin(train['hadm_id'])]
covid_test_imputed = df_test_imputed[df_test_imputed['hadm_id'].isin(test['hadm_id'])]
```

```python
covid_test_imputed.reset_index(drop=True, inplace=True)
covid_train_imputed.reset_index(drop=True, inplace=True)
```

```python
print(f"Size of covid_train_imputed: {covid_train_imputed.shape}")
print(f"Size of covid_test_imputed: {covid_test_imputed.shape}")
```

```python
folderpath = os.path.join(current_dir, OUTPUT_DATA_FOLDER)

covid_train_imputed.to_csv(os.path.join(folderpath, OUTPUT_COVID_TRAIN_DATA_FILE), index=False)
covid_test_imputed.to_csv(os.path.join(folderpath, OUTPUT_COVID_TEST_DATA_FILE), index=False)
```
