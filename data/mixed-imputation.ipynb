# Parameters

```python
## INPUT

MICE_DATA_FOLDER = "mice-col/"
PMM_DATA_FOLDER = "pmm/"

COVID_TRAIN_DATA_FILE = 'patient_covid_train.csv'
COVID_TEST_DATA_FILE = 'patient_covid_test.csv'

CONTROL_PRE_TRAIN_DATA_FILE = 'patient_control_pre_train.csv'
CONTROL_PRE_TEST_DATA_FILE = 'patient_control_pre_test.csv'

CONTROL_POST_TRAIN_DATA_FILE = 'patient_control_post_train.csv'
CONTROL_POST_TEST_DATA_FILE = 'patient_control_post_test.csv'

## OUTPUT

OUTPUT_DATA_FOLDER = "mixed/"
```

# Import Data

```python
import pandas as pd
import os

# Set the path to the datasets
current_dir = os.getcwd()

# MICE-COL

folderpath = os.path.join(current_dir, MICE_DATA_FOLDER)

mice_covid_train = pd.read_csv(os.path.join(folderpath, COVID_TRAIN_DATA_FILE))
mice_covid_test = pd.read_csv(os.path.join(folderpath, COVID_TEST_DATA_FILE))

mice_control_pre_train = pd.read_csv(os.path.join(folderpath, CONTROL_PRE_TRAIN_DATA_FILE))
mice_control_pre_test = pd.read_csv(os.path.join(folderpath, CONTROL_PRE_TEST_DATA_FILE))

mice_control_post_train = pd.read_csv(os.path.join(folderpath, CONTROL_POST_TRAIN_DATA_FILE))
mice_control_post_test = pd.read_csv(os.path.join(folderpath, CONTROL_POST_TEST_DATA_FILE))

# PMM

folderpath = os.path.join(current_dir, PMM_DATA_FOLDER)

pmm_covid_train = pd.read_csv(os.path.join(folderpath, COVID_TRAIN_DATA_FILE))
pmm_covid_test = pd.read_csv(os.path.join(folderpath, COVID_TEST_DATA_FILE))

pmm_control_pre_train = pd.read_csv(os.path.join(folderpath, CONTROL_PRE_TRAIN_DATA_FILE))
pmm_control_pre_test = pd.read_csv(os.path.join(folderpath, CONTROL_PRE_TEST_DATA_FILE))

pmm_control_post_train = pd.read_csv(os.path.join(folderpath, CONTROL_POST_TRAIN_DATA_FILE))
pmm_control_post_test = pd.read_csv(os.path.join(folderpath, CONTROL_POST_TEST_DATA_FILE))
```

```python
print(f"COVID train: {mice_covid_train.shape}, {pmm_covid_train.shape}")
print(f"COVID test: {mice_covid_test.shape}, {pmm_covid_test.shape}")
print(f"Control pre train: {mice_control_pre_train.shape}, {pmm_control_pre_train.shape}")
print(f"Control pre test: {mice_control_pre_test.shape}, {pmm_control_pre_test.shape}")
print(f"Control post train: {mice_control_post_train.shape}, {pmm_control_post_train.shape}")
print(f"Control post test: {mice_control_post_test.shape}, {pmm_control_post_test.shape}") 
```

```python
mixed_covid_train = mice_covid_train.copy()
mixed_covid_test = mice_covid_test.copy()
mice_control_pre_train = mice_control_pre_train.copy()
mice_control_pre_test = mice_control_pre_test.copy()
mice_control_post_train = mice_control_post_train.copy()
mice_control_post_test = mice_control_post_test.copy()

# Replace negative values with PMM values
mixed_covid_train = mixed_covid_train.where(mixed_covid_train >= 0, pmm_covid_train.values)
mixed_covid_test = mixed_covid_test.where(mixed_covid_test >= 0, pmm_covid_test.values)
mixed_control_pre_train = mice_control_pre_train.where(mice_control_pre_train >= 0, pmm_control_pre_train.values)
mixed_control_pre_test = mice_control_pre_test.where(mice_control_pre_test >= 0, pmm_control_pre_test.values)
mixed_control_post_train = mice_control_post_train.where(mice_control_post_train >= 0, pmm_control_post_train.values)
mixed_control_post_test = mice_control_post_test.where(mice_control_post_test >= 0, pmm_control_post_test.values)
```

```python
folderpath = os.path.join(current_dir, OUTPUT_DATA_FOLDER)

mixed_covid_train.to_csv(os.path.join(folderpath, COVID_TRAIN_DATA_FILE))
mixed_covid_test.to_csv(os.path.join(folderpath, COVID_TEST_DATA_FILE))

mixed_control_pre_train.to_csv(os.path.join(folderpath, CONTROL_PRE_TRAIN_DATA_FILE))
mixed_control_pre_test.to_csv(os.path.join(folderpath, CONTROL_PRE_TEST_DATA_FILE))

mixed_control_post_train.to_csv(os.path.join(folderpath, CONTROL_POST_TRAIN_DATA_FILE))
mixed_control_post_test.to_csv(os.path.join(folderpath, CONTROL_POST_TEST_DATA_FILE))
```
